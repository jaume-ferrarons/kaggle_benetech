```python
import pandas as pd
import skimage
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import Image, display

ANNOTATIONS_FILE = "annotations.pkl"
IMAGE_PREFIX = "../data/train/images/"
```

```python
annotations = pd.read_pickle(ANNOTATIONS_FILE)
annotations = annotations[annotations['chart-type'] == 'vertical_bar']
annotations.head()
```

```python
def draw_info(row):
    print(row)
    print(row['axes']['x-axis'])
    tick_labels = [text['text'] for text in row['text'] if text['role'] == 'tick_label']
    image = skimage.io.imread(IMAGE_PREFIX + row['image'])
    plt.imshow(image)
    print(tick_labels)
    x_ticks = np.array([[tick['tick_pt']['x'], tick['tick_pt']['y']] for tick in row['axes']['x-axis']['ticks']])
    y_ticks = np.array([[tick['tick_pt']['x'], tick['tick_pt']['y']] for tick in row['axes']['y-axis']['ticks']])
    y_ticks_values = np.array(list(map(float, tick_labels[:len(y_ticks)])))
    ys = [p['y'] for p in row['data-series']]
    interpolated_tops = np.interp(ys, y_ticks_values[::-1], y_ticks[:,1][::-1])
    plt.scatter(x_ticks[:,0], x_ticks[:,1], c='r')
    plt.scatter(y_ticks[:,0], y_ticks[:,1], c='b')
    plt.scatter(x_ticks[:,0], interpolated_tops, c='y')
    plt.show()
    

for _, row in annotations.sample(10, random_state=0).iterrows():
    draw_info(row)
```
