```python
import easyocr
from IPython.display import Image
import skimage
import matplotlib.pyplot as plt


reader = easyocr.Reader(['en']) # this needs to run only once to load the model into memory
image_path = '../data/train/images/e91e28111e86.jpg'
image_path = '../data/test/images/00f5404753cf.jpg'
result = reader.readtext(image_path)
```

```python
import pytesseract
from PIL import Image


# Load the image
image = Image.open(image_path)

# Convert the image to grayscale
image = image.convert('L')

# Perform OCR using Pytesseract
text = pytesseract.image_to_string(image)

# Print the result
print(text)
```

```python
image = skimage.io.imread(image_path)


plt.imshow(image)
for coord, text, _ in result:
    plt.text((coord[0][0] + coord[1][0])/2, (coord[0][1] + coord[3][1])/2, text, c='r',
                horizontalalignment='center',
                verticalalignment='center',)
plt.show()
```

```python
def display_extraction(img):
    results = reader.readtext(img, mag_ratio=1, text_threshold=0)
    plt.imshow(img)
    for coord, text, _ in results:
        plt.text((coord[0][0] + coord[1][0])/2, (coord[0][1] + coord[3][1])/2, text, c='r',
                    horizontalalignment='center',
                    verticalalignment='center',)
    plt.show()

display_extraction(image[:,:100])
```
